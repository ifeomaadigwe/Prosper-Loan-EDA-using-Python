```python
# Is there a relationship between the borrowers and loan status? 

# Select the borrower rate and loan status columns
borrower_rates = df['BorrowerRate']
loan_status = df['LoanStatus']

# Create a box plot of borrower rate by loan status using Seaborn
sns.boxplot(x=loan_status, y=borrower_rates)

# Set the plot title and axis labels
plt.title("Relationship between Borrowers Rate and Loan Status")
plt.xlabel("Loan Status")
plt.ylabel("Borrowers Rate")

# Rotate the x-axis labels to improve readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

Observation:

With this box plot, we can examine the median, quartiles, and range of rates for each loan status category to identify 
any differences or similarities.-e.g- 
- I observed that loans with a status of "current" tend to have lower median rates compared to loans with a status of "chargedoff".
- I also observed that loans with a status of "defaulted" have a wider range of rates compared to other loan statuses.

```python
# What is the employment status by loan status?

# Group the data by loan status and employment status and count the occurrences
grouped = df.groupby(['LoanStatus', 'EmploymentStatus']).size().reset_index(name='count')

# Pivot the data to make loan status the columns and employment status the index
pivoted = grouped.pivot(index='EmploymentStatus', columns='LoanStatus', values='count')

# Create a stacked bar chart of the data
pivoted.plot(kind='bar', stacked=True)

# Set the plot title and axis labels
plt.title('Employment Status by Loan Status')
plt.xlabel('Employment Status')
plt.ylabel('Customer_count')

# Display the plot
plt.show()
```

Observation:

I observed that the majority of current and completed loans have been issued to borrowers who are employed or have some form of income

```python
df['IsBorrowerHomeowner'].value_counts()
```

```python
# calculate the percentage of borrowers who are homeowners/non-homeowners
home_owner_percentage =  df['IsBorrowerHomeowner'].value_counts(normalize=True) * 100

home_owner_percentage
```

```python
# What is the proportion of homeowners vs non-homeowners

# Calculate the proportion of homeowners and non-homeowners
homeowners_prop = df['IsBorrowerHomeowner'].value_counts(normalize=True)[True]
non_homeowners_prop = df['IsBorrowerHomeowner'].value_counts(normalize=True)[False]

# Create a pie chart of homeowners vs non-homeowners
labels = ['Homeowners', 'Non-Homeowners']
sizes = [homeowners_prop, non_homeowners_prop]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Set the plot title
plt.title('Proportion of Homeowners vs Non-Homeowners')

# Display the plot
plt.show()
```

```python
# Does Borrower been a Homeowner have an impact on loan term length?

sns.set(style="whitegrid")

# group the data by term and borrower homeowner status
grouped = df.groupby(["Term", "IsBorrowerHomeowner"])["IsBorrowerHomeowner"].count().reset_index(name="count")

# calculate the percentage of borrowers who are homeowners/non-homeowners for each term length
grouped["percentage"] = grouped.groupby("Term")["count"].apply(lambda x: x/x.sum()*100)

# create a stacked bar chart
plt.figure(figsize=(10,6))
sns.barplot(x="Term", y="percentage", hue="IsBorrowerHomeowner", data=grouped, palette="Set2")
plt.title("Impact of Borrower Homeownership on Loan Term Length")
plt.xlabel("Loan Term Duration")
plt.ylabel("% of Borrowers")
plt.show()
```

Observation:

- The stacked bar chart shows that the percentage of borrowers who are homeowners is generally higher for longer term loans than borrowers who are 
not homeowners.
- This suggests that homeownership may be a factor in borrowers' decision to choose longer term loans.

```python
# What is the Average number of loan in-take per month?

df['LoanOriginationDate'] = pd.to_datetime(df['LoanOriginationDate'])
start_date = df['LoanOriginationDate'].min()
end_date = df['LoanOriginationDate'].max()

num_months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month) + 1

total_loans = df.shape[0]

avg_loans_per_month = total_loans / num_months

avg_loans_per_month
```

```python
# Show the trend of avg laon per month

# Convert LoanOriginationDate to datetime
df['LoanOriginationDate'] = pd.to_datetime(df['LoanOriginationDate'])

# Group loans by month and calculate average loans per month
monthly_loans = df.groupby(pd.Grouper(key='LoanOriginationDate', freq='M'))['LoanKey'].count()
avg_loans_per_month = monthly_loans.mean()

# Plot line chart
plt.plot(monthly_loans.index, monthly_loans.values)
plt.axhline(y=avg_loans_per_month, color='r', linestyle='-')
plt.title('Average Number of Loans per Month')
plt.xlabel('Month')
plt.ylabel('Number of Loans')
plt.show()
```

```python
# How is the loan original amount related to the monthly loan payment in the Prosper loans data?

sns.scatterplot(x='MonthlyLoanPayment', y='LoanOriginalAmount', data=df)
plt.title('LoanOriginalAmount by MonthlyLoanPayment')
plt.xlabel('Monthly Loan Payment')
plt.ylabel('Loan Original Amount')
plt.show()
```
